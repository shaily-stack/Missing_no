###### Data Transformation

```python
import scipy.stats as stats
```

```python
def made_plots(data, variable):
    plt.figure(figsize=(15,5))
    
    plt.subplot(1,2,1)
    
    data[variable].hist()
    
    plt.subplot(1,2,2)
    stats.probplot(data[variable], dist="norm", plot=plt)
    
    plt.show()
```

```python
made_plots(data, 'price')
```

```python
data['Log_price']=np.log(data['price']+1)
made_plots(data,'Log_price')
```

```python
data['Rec_price']= 1/(data['price']+1)
made_plots(data, 'Rec_price')
```

```python
data['sqr_price']=data['price']**(1/2)
made_plots(data,'sqr_price')
```

```python
data.loc[:,['price','exp_price']].sort_values(by='exp_price')
```

```python
data['exp_price']= np.exp(data['price'])
# made_plots(data, 'exp_price')  We have huge variation exp(0) to exp()->inf
```

###### BOX- COX Transformation

```python
data['price_box_cox'], param = stats.boxcox(data.price +1)

print('Optimal lambda:', param)
made_plots(data, 'price_box_cox')
```
